```python
from sklearn import datasets
from sklearn import manifold
from sklearn.model_selection import train_test_split
from sklearn.cluster import Birch
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.metrics import accuracy_score

import numpy as np

import matplotlib.pyplot as plt

forest = datasets.fetch_covtype(return_X_y=True)
X = forest[0]
y = forest[1]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.005, random_state=42)
print(np.shape(X_test))
print(np.shape(y_test))
```

```python
pca = PCA(2)

df = pca.fit_transform(X_test,y_test)
 
kmeans1 = KMeans(n_clusters= 7)
kmeans2 = KMeans(n_clusters= 100)

label1 = kmeans1.fit_predict(df)
label2 = kmeans2.fit_predict(df)


centroids1 = kmeans1.cluster_centers_
u_labels1 = np.unique(label1)

centroids2 = kmeans2.cluster_centers_
u_labels2 = np.unique(label2)

u_labelsy = np.unique(y_test)
np.shape(u_labelsy)
```

```python
print('measures the similarity of the two assignments with k=7 with dr')
print(metrics.rand_score(y_test, label1))
print(metrics.adjusted_rand_score(y_test, label1))
print('homogeneity: each cluster contains only members of a single class.')
print(metrics.homogeneity_score(y_test, label1))
print('completeness: all members of a given class are assigned to the same cluster')
print(metrics.completeness_score(y_test, label1))
print('accuracy', accuracy_score(y_test, label1))
print('measures the similarity of the two assignments with k=100 with dr')
print(metrics.rand_score(y_test, label2))
print(metrics.adjusted_rand_score(y_test, label2))
print('homogeneity: each cluster contains only members of a single class.')
print(metrics.homogeneity_score(y_test, label2))
print('completeness: all members of a given class are assigned to the same cluster')
print(metrics.completeness_score(y_test, label2))
print('accuracy', accuracy_score(y_test, label2))
```

```python
for i in u_labels1:
    plt.scatter(df[label1 == i , 0] , df[label1 == i , 1] , label = i, s = 5)
    
plt.scatter(centroids1[:,0] , centroids1[:,1] , s = 50, color = 'k', marker= 'x')
plt.show()

for i in u_labels2:
    plt.scatter(df[label2 == i , 0] , df[label2 == i , 1] , label = i, s = 5)
    
plt.scatter(centroids2[:,0] , centroids2[:,1] , s = 50, color = 'k', marker= 'x')
plt.show()
```
