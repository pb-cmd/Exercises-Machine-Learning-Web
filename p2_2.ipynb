```python
from sklearn import datasets
from sklearn import manifold
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.manifold import Isomap

import numpy as np

import matplotlib.pyplot as plt
```

```python

diabetes = datasets.load_diabetes(return_X_y=True)

X = diabetes[0]
y = diabetes[1]
```

```python
lle = manifold.locally_linear_embedding(X,n_neighbors=10,n_components=2)

res=lle[0]
```

```python
plt.scatter( res[:,0],res[:,1], c=y, marker= 'o')
plt.show()
```

```python

pca = PCA(n_components=2)
result2 = pca.fit_transform(X)
```

```python

plt.scatter( result2[:,0] , result2[:,1], c=y ,marker= 'o')
plt.show()
```

```python
t_sne = manifold.TSNE()
sne = t_sne.fit_transform(X)

plt.scatter( sne[:,0] , sne[:,1], c=y, marker= 'o')
plt.show()

```

```python
embedding = Isomap(n_components=2)
iso = embedding.fit_transform(X)
plt.scatter( iso[:,0] , iso[:,1], c=y, marker= 'o')
plt.show()
```
