```python
import networkx as nx
import random as ra
import matplotlib.pyplot as plt
```

```python

er = nx.erdos_renyi_graph(75, 0.15, ra.seed(42))

ba = nx.barabasi_albert_graph(75, 5, ra.seed(42))
nx.draw(er, with_labels=True, font_weight='bold')
plt.plot()
```

```python
nx.draw(ba, with_labels=True, font_weight='bold')
plt.plot
```

```python
erRank = nx.pagerank(er)
erList = list(erRank.values())
x = list(range(0,75))
plt.plot(x, erList)
plt.show()
```

```python
nx.pagerank(ba)
baRank = nx.pagerank(ba)
baList = list(baRank.values())
x = list(range(0,75))
plt.plot(x, baList)
plt.show()
```

```python
# power law bei ba deutlich zu sehen, poisson bei er 
```
